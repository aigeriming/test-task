```python
import pandas as pd #  Pandas    .
import matplotlib.pyplot as plt  #  Matplotlib   .

#  

data = pd.read_csv('weather_data.csv') #    weather_data.csv      data.
print(data.head()) #  5  .
print(data.dtypes) #    .

#  

data.fillna(0, inplace=True) #       0.

#  

average_temp = data['temperature_2m_mean'].mean() #    .
max_temp = data['temperature_2m_max'].max() #    .
min_temp = data['temperature_2m_min'].min() #    .
max_temp_day = data.loc[data['temperature_2m_max'].idxmax(), 'date'] # ,     .
min_temp_day = data.loc[data['temperature_2m_max'].idxmin(), 'date'] # ,     .
total_precipitation = data['precipitation_sum'].sum() #     .
max_precipitation_day = data.loc[data['precipitation_sum'].idxmax(), 'date'] # ,      .

# 

plt.plot(data['date'], data['temperature_2m_mean'], label='   2 ') #     .
plt.plot(data['date'], data['apparent_temperature_mean'], label=' ') #     .
plt.xlabel(' 2023') #  x.
plt.ylabel(' (C)') #  y.
plt.title(' ') # .
plt.legend() #  ..   .
plt.show() # .
print(f"   : {round(average_temp, 1)}C") #    .
print(f"    : {max_temp}C (: {max_temp_day}.05.2023)") #    .
print(f"    : {min_temp}C (: {min_temp_day}.05.2023)") #    .

plt.hist(data['precipitation_sum']) #    .
plt.xlabel(' ()') #  x.
plt.ylabel('') #  y.
plt.title(' ') # .
plt.show() # .
print(f" : {round(total_precipitation, 1)} ") #     .
print(f"    : {max_precipitation_day}.05.2023") # ,      .
```
