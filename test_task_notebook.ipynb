```python
import pandas as pd #импортирует библиотеку Pandas для работы с данными.
import matplotlib.pyplot as plt  #импортирует библиотеку Matplotlib для построения графиков.

#ОЗНКОМЛЕНИЕ С ДАННЫМИ

data = pd.read_csv('weather_data.csv') #считывает данные из файла weather_data.csv и сохраняет их в переменную data.
print(data.head()) #выводит первые 5 строк данных.
print(data.dtypes) #выводит типы данных в столбцах.

#ПРЕДВАРИТЕЛЬНАЯ ОБРАБОТКА ДАННЫХ

data.fillna(0, inplace=True) #заменяет все пропущенные значения в данных на 0.

#ОСНОВНОЙ АНАЛИЗ ДАННЫХ

average_temp = data['temperature_2m_mean'].mean() #вычисляет среднюю температуру за месяц.
max_temp = data['temperature_2m_max'].max() #находит максимальную температуру за месяц.
min_temp = data['temperature_2m_min'].min() #находит минимальную температуру за месяц.
max_temp_day = data.loc[data['temperature_2m_max'].idxmax(), 'date'] #находит дату, когда была зафиксирована максимальная температура.
min_temp_day = data.loc[data['temperature_2m_max'].idxmin(), 'date'] #находит дату, когда была зафиксирована минимальная температура.
total_precipitation = data['precipitation_sum'].sum() #вычисляет общее количество осадков за месяц.
max_precipitation_day = data.loc[data['precipitation_sum'].idxmax(), 'date'] #находит дату, когда было зафиксировано наибольшее количество осадков.

#ВИЗУАЛИЗАЦИЯ ДАННЫХ

plt.plot(data['date'], data['temperature_2m_mean'], label='температура на высоте 2 метра') #строит график средней температуры за месяц.
plt.plot(data['date'], data['apparent_temperature_mean'], label='ощущаемая температура') #строит график ощущаемой температуры за месяц.
plt.xlabel('Май 2023') #подписывает ось x.
plt.ylabel('Температура (°C)') #подписывает ось y.
plt.title('Изменения температуры') #подписывает график.
plt.legend() #добавляет легенду т.е. названия линий графиков.
plt.show() #отображает график.
print(f"Средняя температура за месяц: {round(average_temp, 1)}°C") #выводит среднюю температуру за месяц.
print(f"Самая высокая температура за месяц: {max_temp}°C (дата: {max_temp_day}.05.2023)") #выводит максимальную температуру за месяц.
print(f"Самая низкая температура за месяц: {min_temp}°C (дата: {min_temp_day}.05.2023)") #выводит минимальную температуру за месяц.

plt.hist(data['precipitation_sum']) #строит гистограмму осадков за месяц.
plt.xlabel('Осадки (мм)') #подписывает ось x.
plt.ylabel('Частота') #подписывает ось y.
plt.title('Частота осадков') #подписывает график.
plt.show() #отображает график.
print(f"Всего осадков: {round(total_precipitation, 1)} мм") #выводит общее количество осадков за месяц.
print(f"День с наибольшими выпавшими осадками: {max_precipitation_day}.05.2023") #выводит дату, когда было зафиксировано наибольшее количество осадков.
```
